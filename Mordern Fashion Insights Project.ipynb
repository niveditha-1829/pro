## Fashion Characteristics in Modern Society

This project analyzes fashion preferences and behaviors in modern society using survey data. It explores how age, gender, profession, and lifestyle influence clothing choices, style preferences, and shopping habits.

 Objectives:
- Identify popular styles, colors, and shopping patterns
- Understand the link between fashion and personality
- Analyze influences like social media, functionality, and experimentation


## Importing Labraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Load Dataset

```python
df = pd.read_excel("Fashion(Data Points) .xlsx", sheet_name="Fashion(Data Points) - Form res")
df.head()
```

```python
df.info()
df.describe(include='all')
df.isnull().sum()
df.duplicated().sum()
```

```python
df.dropna(how='all', inplace=True)
df['Style reflects personality (1–10)'].fillna(df['Style reflects personality (1–10)'].mean(), inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('–', '-')
```

## Gender Distribution

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Gender Distribution', ylabel='')
plt.show()
```

## Age Group Count

```python
sns.countplot(data=df, x='Age_Group')
plt.title('Age Group Distribution')
plt.xticks(rotation=45)
plt.show()
```

## Most Preferred Style

```python
df['Style_Preferences'].value_counts().plot(kind='bar', title='Preferred Fashion Styles')
plt.ylabel('Number of People')
plt.show()
```

## Favorite Color Palette

```python
df['Favorite_color_palette_for_clothing'].value_counts().plot(kind='barh', title='Favorite Color Palettes')
plt.xlabel('Count')
plt.show()
```

## Shopping Frequency

```python
sns.countplot(data=df, y='Shopping', order=df['Shopping'].value_counts().index)
plt.title('Fashion Shopping Frequency')
plt.show()
```

## Influence on Clothing Purchases

```python
df['Influences_your_clothing_purchases'].value_counts().plot(kind='barh')
plt.title('What Influences Clothing Purchases')
plt.xlabel('Respondents')
plt.show()
```

## Aesthetic vs Functionality Preference

```python
df['Functionality_or_aesthetics'].value_counts().plot(kind='pie', autopct='%1.1f%%', ylabel='', title='Functionality vs Aesthetics Preference')
plt.show()
```

## Footwear Preference

```python
df['Footwear'].value_counts().plot(kind='bar', title='Footwear Preferences')
plt.ylabel('Count')
plt.show()
```

## Wardrobe Type

```python
df['Describes_your_wardrobe'].value_counts().plot(kind='barh', title='Wardrobe Types')
plt.xlabel('Number of Respondents')
plt.show()
```

## Typical Clothing Shopping Locations

```python
df['Typically_shop_for_clothes'].value_counts().plot(kind='bar', title='Preferred Shopping Locations')
plt.ylabel('People')
plt.show()
```

## Experimentation with New Styles

```python
sns.countplot(data=df, x='Experiment_with_new_styles', order=df['Experiment_with_new_styles'].value_counts().index)
plt.title('Style Experimentation Frequency')
plt.xticks(rotation=30)
plt.show()
```

## Gender vs Fashion Style Preference

```python
sns.countplot(data=df, x='Style_Preferences', hue='Gender')
plt.title('Style Preferences by Gender')
plt.xticks(rotation=45)
plt.show()
```

## Age Group vs Shopping Frequency

```python
pd.crosstab(df['Age_Group'], df['Shopping']).plot(kind='bar', stacked=True)
plt.title('Shopping Frequency by Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Style preferences are most common across different age groups

```python
plt.figure(figsize=(12,6))
style_by_age = pd.crosstab(df['Age_Group'], df['Style_Preferences'])
style_by_age.plot(kind='bar', stacked=True, colormap='Set3')
plt.title('Style Preferences by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.legend(title='Style Preferences', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Which factors most influence fashion purchases

```python
plt.figure(figsize=(10,5))
df['Influences_your_clothing_purchases'].value_counts().plot(kind='barh', color='mediumseagreen')
plt.title('Top Factors Influencing Clothing Purchases')
plt.xlabel('Number of Respondents')
plt.ylabel('Influencing Factor')
plt.grid(True)
plt.show()
```

## Conclusion:


This analysis reveals how fashion in modern society is shaped by a combination of personal preferences, lifestyle factors, and social influences. We observed clear patterns in style choices based on age, gender, and profession, as well as the impact of experimentation and personality alignment on clothing behavior.

Key takeaways include:
- Casual and minimalist styles are most common across age groups.
- Many individuals feel their fashion style reflects their personality.
- Factors like comfort, trend influence, and social media strongly guide purchasing decisions.
